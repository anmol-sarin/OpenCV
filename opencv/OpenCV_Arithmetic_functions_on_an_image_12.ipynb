## merging two images

blended = cv2.addWeighted(src1, alpha, src2, beta, gamma)

	• src1: First image
	• alpha: Weight of the first image (intensity)
	• src2: Second image
	• beta: Weight of the second image
	• gamma: Scalar added to each sum


```python
img1 = cv2.imread(r'C:\Users\anmol sarin\OneDrive\Desktop\me\opencv\images\img-1.webp')
img2 = cv2.imread(r'C:\Users\anmol sarin\OneDrive\Desktop\me\opencv\images\img-2.webp')
img1_resize = cv2.resize(img1,(500,500))
img2_resize = cv2.resize(img2,(500,500))
new_img = cv2.addWeighted(img1_resize , 1 , img2_resize , .5,0)
cv2.imshow("merged image" , new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
