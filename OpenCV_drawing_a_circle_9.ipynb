## Drawing Circle on an image using OpenCV

```python
import cv2
```

```python
Old_img = cv2.imread(r"C:\Users\anmol sarin\OneDrive\Desktop\me\opencv\images\img-3.webp")
img_resize = cv2.resize(Old_img,(400,400))
new_img = cv2.circle(img = img_resize,center=[200,150],radius = 87,color=[0,255,0],thickness = 2, lineType = 16)
cv2.imshow("image" , new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
Old_img = cv2.imread(r"C:\Users\anmol sarin\OneDrive\Desktop\me\opencv\images\img-3.webp")
img_resize = cv2.resize(Old_img,(400,400))
new_img = cv2.circle(img = img_resize,center=[200,150],radius = 80,color=[0,255,0],thickness = -1, lineType = 16)
cv2.imshow("image" , new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
